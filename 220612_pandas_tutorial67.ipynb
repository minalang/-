## 06 Pivot
활용데이터<br>
국가별 5세이하 사망비율 통계 : https://www.kaggle.com/utkarshxy/who-worldhealth-statistics-2020-complete <br>
Dataurl = https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv')
df.head()
```

### 83. Indicator을 삭제하고 First Tooltip 컬럼에서 신뢰구간에 해당하는 표현을 지워라

```python
#indicator열 지우기
df.drop('Indicator', axis = 1, inplace = True)

# first tooltip에서 신뢰구간에 해당하는 표현 지우기
df['First Tooltip'] = df['First Tooltip'].map(lambda x: float(x.split('[')[0]))
df.head()
```

84. 년도가 2015년 이상, Dim1이 Both sexes인 케이스만 추출하라

```python
df.dtypes
```

```python
ans84 = df[(df['Period'] >= 2015) & (df['Dim1'] == 'Both sexes')]
ans84.head()
```

85. 84번 문제에서 추출한 데이터로 아래와 같이 나라에 따른 년도별 사망률을 데이터 프레임화 하라

```python
ans85 = ans84.pivot(index = 'Location', columns = 'Period', values = 'First Tooltip')
ans85.head()
```

### 86. Dim1에 따른 년도별 사망비율의 평균을 구하라

```python
ans86 = df.pivot_table(index = 'Dim1', columns = 'Period', values = 'First Tooltip', aggfunc = 'mean')
ans86.iloc[:, :4]
```

활용데이터<br>

87. 데이터에서 한국 KOR 데이터만 추출하라

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/winter.csv')
df.head()
```

```python
ans87 = df[df['Country'] == 'KOR']
ans87.head()
```

88. 한국 올림픽 메달리스트 데이터에서 년도에 따른 medal 갯수를 데이터프레임화 하라

```python
ans88 = ans87.pivot_table(index = 'Year', columns = 'Medal', aggfunc = 'size').fillna(0)
ans88.head()
```

89. 전체 데이터에서 sport종류에 따른 성별수를 구하여라

```python
ans89 = df.pivot_table(index = 'Sport', columns = 'Gender', aggfunc = 'size').fillna(0)
ans89
```

90. 전체 데이터에서 Discipline종류에 따른 따른 Medal수를 구하여라

```python
ans90 = df.pivot_table(index = 'Discipline', columns = 'Medal', aggfunc = 'size')
ans90
```

## 07 Merge, Concat

활용데이터<br>
국가별 5세이하 사망비율 통계 : https://www.kaggle.com/utkarshxy/who-worldhealth-statistics-2020-complete<br>
(데이터 변형)<br>
Dataurl = https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
df
```

```python
# 데이터 분리
df1 = df.iloc[:4, :]
df2 = df.iloc[4:, :]
display(df1)
display(df2)
```

### 91. df1과 df2 데이터를 하나의 데이터 프레임으로 합쳐라

```python
ans91 = pd.concat([df1, df2], axis = 0, ignore_index = False)
ans91
```

### 92. df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 둘다 포함하고 있는 년도에 대해서만 고려한다

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
ans92 = pd.concat([df3, df4], join = 'inner')
ans92
```

93. df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 모든 컬럼을 포함하고, 결측치는 0으로 대체한다

```python
ans93 = pd.concat([df3, df4], join = 'outer').fillna(0)
ans93
```

94. df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합쳐라. Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력하라

```python
df.T
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
ans94 = df5.merge(df6, how = 'inner', on = 'Algeria')
ans94
```

95. df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합쳐라. Algeria컬럼을 key로 하고 합집합으로 합쳐라

```python
ans95 = df5.merge(df6, how = 'outer', on = 'Algeria')
ans95
```
