### 32. df의 item_name컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱싱

### 33. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

```python
#중복행 제거 'drop_duplicates'
ans33 = df[(df['item_name'] == 'Steak Salad') | (df['item_name'] == 'Bowl')]
ans33.drop_duplicates('item_name')
```

### 34. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라

```python
#중복행 제거 'drop_duplicates', 마지막 남기기 keep = 'last'
ans33 = df[(df['item_name'] == 'Steak Salad') | (df['item_name'] == 'Bowl')]
ans33.drop_duplicates('item_name', keep = 'last')
```
