## 문제1: 데이터 정렬과 평균치 산출
alchol(음주수치)이 상위 10번째 값으로 상위 1~10위의 범위의 값을 변경 후 speeding(속도)가 7이상의 alchol데이터들의 평균 산출
최종결과만 print

```python
import seaborn as sns
import pandas as pd
sns.get_dataset_names()
df = sns.load_dataset('car_crashes')
df.head()
```

```python
#alchol열 값으로 데이터 정렬(내림차순)
sub1 = df.sort_values('alcohol', ascending = False)
#10번째의 값 출력
ten = sub1.iloc[9, 2]
#데이터 변환
sub1.iloc[0:9, 2] = ten
#조건에 따라 데이터 간추림
sub2 = sub1[sub1['speeding'] >= 7]
#결과 출력
print(f'최종결과물은: {sub2["alcohol"].mean()}')
```

## 문제2: 특정 비율만큼의 데이터와 결측치 대체, 전처리 전후 표준편차 전후

데이터 첫번째 행부터 70%의 데이터를 추출한 후<br> 
distance(거리)의 결측값을 distance(거리)의 중앙값으로 대체하고<br>
결측값 전처리 전과 후의 표준편차를 비교<br>
전/후 결과값 print 출력

```python
import numpy as np
df = sns.load_dataset('planets')
```

```python
#데이터 70%추출하기
data_len = int(len(df)*70)
df = df[:data_len]
df2 = df.copy()

#distance의 중앙값 구하기
df2['distance'] = df2['distance'].fillna(df['distance'].median())

#표준편차 구하기
std1 = np.std(df['distance'])
std2 = np.std(df2['distance'])

print(f'결측값 전처리 전 표준편차: {std1}\n 결측값 전처리 후 표준편차: {std2}')
```

## 문제3: 통계기법(IQR) 방식의 이상치 제거
orbital_period열의 이상치를 IQR기법으로 제거<br>
이상치들의 합을 계산<br>
최종결과를 print


```python
import seaborn as sns
sns.get_dataset_names()
df = sns.load_dataset('planets')
df.head()
```

```python
#quantile값 구하기
q25 = df['orbital_period'].quantile(0.25) #25% 수준의 값
q75 = df['orbital_period'].quantile(0.75) #75% 수준의 값

#IQR값 구하기
iqr = q75 - q25

#min값, max값 구하기
min = q25 - iqr*1.5
max = q75 + iqr*1.5

#이상치 데이터 추출하기
df_outlier = df[(df['orbital_period'] < min) | (df['orbital_period'] > max)]
print(f"정답은 {df_outlier['orbital_period'].sum()}이다")
```

## 문제1: 데이터 전처리


```python
import seaborn as sns
import pandas as pd
import numpy as np
df = sns.load_dataset('penguins')
df.info()
```

```python
df.head()
```

```python
df.isna().sum()
```

연속형 변수: 중앙값, 평균<br>
범주형 변수: 비율이 많은 값으로 채워주는 경우

```python
df['sex'].value_counts()
```

### 1. 결측치 제거

```python
missing = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
for i in missing:
  df[i] = df[i].fillna(df[i].median())
df['sex'] = df['sex'].fillna('Male')
df.info()
```

```python
df.isna().sum() #결측치 제거
```

### 2. 라벨 인코딩

```python
#라벨인코딩
from sklearn.preprocessing import LabelEncoder
label = ['species', 'island', 'sex']
df[label] = df[label].apply(LabelEncoder().fit_transform)
```

```python
df.head()
```

### 3. 데이터 변환, 더미처리

```python
import pandas as pd
category = ['island', 'sex']
for i in category:
  df[i] = df[i].astype('category')
df = pd.get_dummies(df)
df.head()
```

### 4. 파생변수 만들기

```python
#기존 변수 가공하여 새로운 변수 만들기
#qcut: 연속형 데이터의 구간을 나눠 새로운 범주형 변수로 나타내주는 것
df['body_mass_g_qcut'] = pd.qcut(df['body_mass_g'], 5, labels = False)
df.head()
```

### 5. 스케일링

```python
#스케일링 0~1사이의 값으로 맞추어줌
from sklearn.preprocessing import MinMaxScaler
scaler = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
min = MinMaxScaler()
min.fit(df[scaler])
df[scaler] = min.transform(df[scaler])
df.head()
```

### 6. 데이터 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df['species'], test_size = 0.2, stratify = df['species'], random_state = 777)
```
