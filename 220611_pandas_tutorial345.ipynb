### 58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'
```

```python
df['newIncome'] = df.Income_Category.apply(changeCategory)
df.head()
```

### 59. Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라.<br> 
 각 구간의 빈도수를 출력하라
 (0~9 : 0 , 10~19 :10 , 20~29 :20

```python
def age(x):
  if 0 <= x and x<=9:
    return 0
  elif 10 <= x and x<=19:
    return 10
  elif 20 <= x and x<=29:
    return 20
  elif 30 <= x and x<=39:
    return 30
  elif 40 <= x and x<=49:
    return 40
  elif 50 <= x and x<=59:
    return 50
  elif 60 <= x and x<=69:
    return 60
  elif 70 <= x and x<=79:
    return 70
```

```python
df['AgeState'] = df['Customer_Age'].apply(age)
df.groupby('AgeState').size()
```

```python
df['AgeState'] = df.Customer_Age.map(lambda x: x//10 * 10)
df['AgeState'].value_counts().sort_index()
```

### 60. Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라


```python
df['newEduLevel'] = df.Education_Level.map(lambda x:1 if 'Graduate' in x else 0)
df['newEduLevel'].value_counts().sort_index()
```

61. Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라. newLimit 각 값들의 빈도수를 출력하라

```python
df['newLimit'] = df.Credit_Limit.apply(lambda x:1 if x >= 4500 else 0)
df['newLimit'].value_counts().sort_index()
```

### 62. Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라

```python
def check(x):
  if x.Marital_Status == 'Married' and x.Card_Category == 'Platinum':
    return 1
  else: 
    return 0

df['newState'] = df.apply(check, axis = 1)
df.newState.value_counts()
```

63. Gender 컬럼값 M인 경우 male F인 경우 female로 값을 변경하여 Gender 컬럼에 새롭게 정의하라. 각 value의 빈도를 출력하라

```python
def fix_century(x):
  import datetime
  year = x.year - 100 if x.year >= 2061 else x.year
  return pd.to_datetime(datetime.date(year, x.month, x.day))
df['Yr_Mo_Dy'] = df.Yr_Mo_Dy.apply(fix_century)
df.head()
```

### 76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다
일반 object변수의 형변환 심화

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(":")[0]
    
    if hour =='24':
        hour ='00:00:00'
        
        FinalDate = pd.to_datetime(date +" "+hour) +datetime.timedelta(days=1)
        
    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date +" "+hour)
    
    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

df.head()
```
